```julia
import Pkg; 

if split(pwd(),"/")[end] == "_paper_2"
    cd(joinpath(@__DIR__, "../../"))
    Pkg.activate("Project.toml")
end

using CairoMakie
using JLD2
```

```julia
pt_per_in = 72
width = 3.42
height = 5.2
f_size = (pt_per_in * width, pt_per_in * height)
border_width = 0.5
pad = 2

f = Figure(size = f_size, fontsize = 12)
png_dimension = f_size[1] * 0.4

ga = f[1, 1] = GridLayout()
img = load("notebooks/_paper_2/pngs/1/6r7m_assembled.png")
ax, _ = image(ga[1, 1], rotr90(img),
    axis = (
        width = png_dimension, 
        height = png_dimension, 
        title = "",
        spinewidth=border_width
        )
    )

hidespines!(ax)
hidedecorations!(ax)

gb = f[1, 2] = GridLayout()
img = load("notebooks/_paper_2/pngs/1/2tmv_min.png")
ax, _ = image(gb[1, 1], rotr90(img),
    axis = (
        width = png_dimension, 
        height = png_dimension, 
        title = "",
        spinewidth=border_width
        )
    )

hidespines!(ax)
hidedecorations!(ax)

gc = f[2, 1] = GridLayout()
img = load("notebooks/_paper_2/pngs/1/6sae_2.png")
ax, _ = image(gc[1, 1], rotr90(img),
    axis = (
        width = png_dimension, 
        height = png_dimension, 
        title = "",
        spinewidth=border_width
        )
    )
hidespines!(ax)
hidedecorations!(ax)

gd = f[2, 2] = GridLayout()
img = load("notebooks/_paper_2/pngs/1/6sag_v.png")
ax, _ = image(gd[1, 1], rotr90(img),
    axis = (
        width = png_dimension, 
        height = png_dimension, 
        title = "",
        spinewidth=border_width
        )
    )
hidespines!(ax)
hidedecorations!(ax)

ge = f[3, 1] = GridLayout()
img = load("notebooks/_paper_2/pngs/1/6r7m_yin_yang.png")
ax, _ = image(ge[1, 1], rotr90(img),
    axis = (
        width = png_dimension, 
        height = png_dimension, 
        title = "",
        spinewidth=border_width
        )
    )
hidespines!(ax)
hidedecorations!(ax)

gf = f[3, 2] = GridLayout()
img = load("notebooks/_paper_2/pngs/1/1ei7_min.png")
ax, _ = image(gf[1, 1], rotr90(img),
    axis = (
        width = png_dimension, 
        height = png_dimension, 
        title = "",
        spinewidth=border_width
        )
    )
hidespines!(ax)
hidedecorations!(ax)

for (label, layout) in zip(["A", "B", "C", "D", "E", "F"], [ga, gb, gc, gd, ge, gf])
    Label(layout[1, 1, TopLeft()], label,
        fontsize = 12,
        font = :bold,
        padding = (0, pad, pad, 0),
        halign = :right)
end

rowsize!(f.layout, 1, Relative(0.1))
rowgap!(f.layout, 0)

rs = 0.34
colsize!(f.layout, 1, Relative(0.5))
colsize!(f.layout, 2, Relative(0.5))
rowsize!(f.layout, 1, Relative(rs))
rowsize!(f.layout, 2, Relative(rs))
rowsize!(f.layout, 3, Relative(rs))

save(string("notebooks/_paper_2/output/low_energy_states.pdf"), f, pt_per_unit = 1)
f
```

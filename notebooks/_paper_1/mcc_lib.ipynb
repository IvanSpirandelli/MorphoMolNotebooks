```julia
function ply_to_configuration(filepath)
    coordinate_data = readlines(filepath)[12:end]
    flat_realization = Vector{Float64}([])
    radii = Vector{Float64}([])
    # Assuming lines are formatted like "i: x y z r whatever"
    for line in coordinate_data
        line_array = split(line, " ")
        push!(flat_realization, parse(Float64, line_array[1]))
        push!(flat_realization, parse(Float64, line_array[2]))
        push!(flat_realization, parse(Float64, line_array[3]))
        push!(radii, parse(Float64, line_array[4]))
    end
    flat_realization, radii
end

ply_to_configuration("assets/mcc_cluster_plys/4_ball_lib/tet.ply")
```

```julia
mcc_configurations = []
for (root, dirs, files) in walkdir("assets/mcc_cluster_plys/")
    for file in files
        filepath = joinpath(root, file)
        name = split(file, ".")[1]
        if name != ""
            realization, radii = ply_to_configuration(filepath)
            push!(mcc_configurations, (name = name, centers = realization, radii = radii))
        end
    end
end
```
